## Import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

## Load the dataset

```python
train = pd.read_csv("Property_Price_Train.csv")
test = pd.read_csv("Property_Price_Test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
# Remove Id column from train and test dataset
train = train.drop(['Id'], axis=1)
test = test.drop(['Id'], axis=1)
```

```python
# Concatenate train and test file
alldata = pd.concat([train, test], axis=0)
```

```python
# Reset index
alldata = alldata.reset_index()
```

```python
alldata.head()
```

## Dimensions of dataset

```python
alldata.shape
```

```python
# check the datatype of variables
alldata.dtypes
```

```python
# Basic description of data
alldata.describe()
```

## Devide the dataset into numerical and categorical features

```python
df_numeric_features = alldata.select_dtypes(include=[np.number])
df_numeric_features.columns
```

```python
df_categorical_features = alldata.select_dtypes(include=[object])
df_categorical_features.columns
```

## Missing Values Analysis

```python
missing = alldata.isnull().sum()
missing
```

## Missing value treatment for numerical features


```python
# fill na with median
df_numeric_features.fillna(df_numeric_features.median(), inplace=True)
```

```python
# Recheck null values
df_numeric_features.isnull().sum()
```

```python
df_numeric_features.corr()
```

## Missing value treatment for categorical features

```python
# fill na with None
df_categorical_features.fillna('None', inplace=True)
```

```python
# Recheck null values
df_categorical_features.isnull().sum()
```

```python
# Extract Depnding Variable
y = df_numeric_features['Sale_Price']
y.head()
```

```python
y.shape
```

```python
df_numeric_features.head()
```

## Scale our numeric variables using min-max normalization

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
df_housingdata_numcols = pd.DataFrame(min_max.fit_transform(df_numeric_features.iloc[:,0:36]),
                                      columns = df_numeric_features.iloc[:,0:36].columns.tolist())
```

```python
df_housingdata_numcols.head()
```

## Perform one-hot encoding on categorical variables

```python
one_hot_encoded_variables = pd.get_dummies(df_categorical_features, drop_first=True)
```

```python
one_hot_encoded_variables.head()
```

## Concatenate the numeric and encoded variables to the dataframe

```python
df_housedata = pd.concat([df_housingdata_numcols,one_hot_encoded_variables], axis=1)
```

```python
df_housedata.head()
```

```python
# Extract Independent Variables
x = df_housedata
x.head()
```

## Create training and testing datasets using the train_test_split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.10, random_state=10)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

lr =  LinearRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
print(lr.score(x_test,y_test))
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn import metrics
lr_MAE = metrics.mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error (MAE):', lr_MAE)

# calculate MSE
lr_MSE = metrics.mean_squared_error(y_test, y_pred)
print('Mean Squared Error (MSE):', lr_MSE)

# calculate RMSE
lr_RMSE = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error (RMSE):', lr_RMSE)
```

## Decision Tree Model

```python
from sklearn.tree import DecisionTreeRegressor

dt =  DecisionTreeRegressor()
dt.fit(x_train, y_train)
dt_pred = dt.predict(x_test)
print(dt.score(x_test,y_test))
```

```python
y_pred = dt.predict(x_test)
```

```python
dt_MAE = metrics.mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error (MAE):', dt_MAE)

# calculate MSE
dt_MSE = metrics.mean_squared_error(y_test, y_pred)
print('Mean Squared Error (MSE):', dt_MSE)

# calculate RMSE
dt_RMSE = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error (RMSE):', dt_RMSE)
```

## Random Forest Regression Model

```python
from sklearn.ensemble import RandomForestRegressor

rf =  RandomForestRegressor()
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
print(rf.score(x_test,y_test))
```

```python
y_pred = rf.predict(x_test)
```

```python
rf_MAE = metrics.mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error (MAE):', rf_MAE)

# calculate MSE
rf_MSE = metrics.mean_squared_error(y_test, y_pred)
print('Mean Squared Error (MSE):', rf_MSE)

# calculate RMSE
rf_RMSE = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error (RMSE):', rf_RMSE)
```

## Hyper parameter tunning with grid serch cv


```python
paramaters = [{'n_estimators': [ 85, 100],
                     'min_samples_split': [15,20],
                     'max_depth': [8, 10],
                     'min_samples_leaf': [5,10],
                     'max_leaf_nodes': [10, 15]
                    }]
```

```python
from sklearn.model_selection import GridSearchCV
rf_reg_CV= RandomForestRegressor(random_state = 10)
grid = GridSearchCV(estimator = rf_reg_CV,
                         param_grid = paramaters,
                         cv = 5)
rf_grid = grid.fit(x_train, y_train)

# get the best parameters
print('Best parameters for random forest regressor: ', rf_grid.best_params_, '\n')
```

## Build the model using the hyperparameters

```python
rf_grid =  RandomForestRegressor(n_estimators=85,
    max_depth=8,
    min_samples_split=15,
    min_samples_leaf=10,
    max_leaf_nodes=15,
    )
rf_grid.fit(x_train, y_train)
rf_grid_pred = rf_grid.predict(x_test)
print(rf_grid.score(x_test,y_test))
```
